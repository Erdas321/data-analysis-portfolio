```python
# IMPORTING REQUIRED LIBRARIES
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# READING AND STORING DATASET INTO df VARIABLE
```

```python
df = pd.read_csv(r"ccf_sampple.csv")
df
```

```python
# SPLITTING DATASET INTO TRAIN (70%) AND TEST(30%) USING RANDOM SAMPLING
# THE test DATASET SAMPLE WILL BE USED IN ALL SUBSEQUENT OPERATIONS FROM NOW ON
```

```python
train, test = train_test_split(df, test_size = 0.3, random_state = 42)
test.info()
```

```python
# CHANGING MAXIMUM AMOUNT OF DISPLAYED ROWS AND COLUMNS
```

```python
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 100000)
```

```python
# STANDARDIZING gender COLUMN (F TO FEMALE, M TO MALE)
```

```python
test.loc[test['gender'] == 'F', 'gender'] = 'Female'
test.loc[test['gender'] == 'M', 'gender'] = 'Male'
test
```

```python
# NUMBER OF FRAUDULENT TRANSACTIONS PER STATE
```

```python
test[test['is_fraud'] == 1]['state'].value_counts().reset_index().groupby('count').agg({'state': lambda x: ', '.join(x)}).sort_values(by = 'count', ascending = False)
```

```python
# NUMBER OF FRAUDULENT TRANSACTIONS PER CITY
```

```python
test[test['is_fraud'] == 1]['city'].value_counts().reset_index().groupby('count').agg({'city': lambda x: ', '.join(x)}).sort_values(by = 'count', ascending = False)
```

```python
# DISPLAYING FRAUDULENT TRANSACTION DENSITY USING A SCATTER PLOT
```

```python
test[test['is_fraud'] == 1].plot(
    kind = 'scatter',
    x = 'long',
    y = 'lat',
    grid = True,
    label = 'fraudulent transactions',
    s = 50,
    c = test[test['is_fraud'] == 1]['amt'],
    cmap = 'viridis',
    colorbar = True,
    legend = True,
    sharex = False,
    figsize = (10, 7),
    alpha = 0.2
)
plt.show()
```

```python
# DISPLAYING FRAUDULENT TRANSACTIONS PER CATEGORY USING A HORIZONTAL BAR PLOT
```

```python
test[test['is_fraud'] == 1]['category'].value_counts().sort_values(ascending = True).plot(kind = 'barh')
```

```python
# PERCENTAGE OF DATASET POPULATION INVOLVED IN TRANSACTIONAL FRAUD
```

```python
city_pop_full = test.groupby('city')['city_pop'].sum().reset_index().copy()
city_pop_fraud = test[test['is_fraud'] == 1].groupby('city')['city_pop'].sum().reset_index().copy()
city_pop_merge = city_pop_full.merge(city_pop_fraud, how = 'inner', on = 'city', suffixes = ('_full', '_fraud')).copy()
city_pop_merge['city_pop_fraud_percentage'] = (city_pop_merge['city_pop_fraud'] / city_pop_merge['city_pop_full']) * 100
city_pop_merge.sort_values(by = 'city_pop_fraud_percentage', ascending = False).head(20)
```

```python
# DISPLAYING FRAUDULENT TRANSACTIONS PER MONTH USING A HORIZONTAL BAR PLOT
```

```python
test['trans_date_trans_time'] = pd.to_datetime(test['trans_date_trans_time'])
test[test['is_fraud'] == 1]['trans_date_trans_time'].dt.month_name().value_counts().plot(kind = 'barh', ylabel = 'Month', xlabel = 'Fraud Count', grid = True)
```

```python
# DISPLAYING FRAUDULENT TRANSACTIONS PER DAY USING A HORIZONTAL BAR PLOT
```

```python
test[test['is_fraud'] == 1]['trans_date_trans_time'].dt.day_name().value_counts().plot(kind = 'barh', ylabel = 'Week Day', xlabel = 'Fraud Count', grid = True)
```

```python
# TOP REPEAT OFFENDER NAMES IN FRAUDULENT ACTIVITY
```

```python
test['full_name'] = test['first'] + ' ' + test['last']
test[test['is_fraud'] == 1]['full_name'].value_counts().reset_index().groupby('count').agg({'full_name': lambda x: ', '.join(x)}).sort_values(by = 'count', ascending = False)
```
