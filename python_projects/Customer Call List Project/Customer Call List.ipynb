```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re  # Regular expressions for text cleaning
```

```python
# CLEANING DATASET
```

```python
# Load the Excel file into a pandas DataFrame
df = pd.read_excel(r"customer_data.xlsx")
df
```

```python
# Clean 'Last_Name' column by removing non-alphabetic characters
last_name = df['Last_Name']
last_name = list(last_name)
last_name = list(map(lambda x: re.sub(r'[^a-zA-Z]+', '', str(x)), last_name))  # Remove non-alphabetic characters
last_name = list(map(lambda x: re.sub(r'nan', '', str(x)), last_name))  # Remove 'nan' values
df['Last_Name'] = last_name
df
```

```python
# Clean 'Phone_Number' column by keeping only numeric digits
new_numbers = df['Phone_Number']
new_numbers = list(map(lambda x: re.sub(r'[^0-9]+', '', str(x)), new_numbers))  # Remove non-numeric characters
df['Phone_Number'] = new_numbers
df
```

```python
# Split 'Address' column into separate components (Street, State, Zipcode)
new_address = df['Address']
new_address = list(map(lambda x: x.split(','), new_address))  # Split address by commas
df['address'] = list(map(lambda x: x[0], new_address))  # Extract street address
df['state'] = list(map(lambda x: x[1] if len(x) > 1 else '', new_address))  # Extract state
df['zipcode'] = list(map(lambda x: x[2] if len(x) > 2 else '', new_address))  # Extract zipcode
df
```

```python
# Drop a column that is not useful
df = df.drop('Not_Useful_Column', axis=1)
df
```

```python
# Standardize 'Paying Customer' column values ('Y' -> 'Yes', 'N' -> 'No')
customers = df['Paying Customer']
customers = list(map(lambda x: re.sub(r'^Y$', 'Yes', x), customers))
customers = list(map(lambda x: re.sub(r'^N$', 'No', x), customers))
customers = [x if x == 'Yes' or x == 'No' else '' for x in customers]  # Remove invalid values
df['Paying Customer'] = customers
df
```

```python
# Standardize 'Do_Not_Contact' column values ('Y' -> 'Yes', 'N' -> 'No')
contacts = df['Do_Not_Contact']
contacts = list(map(lambda x: re.sub(r'^Y$', 'Yes', str(x)), contacts))
contacts = list(map(lambda x: re.sub(r'^N$', 'No', str(x)), contacts))
contacts = [x if x == 'Yes' or x == 'No' else '' for x in contacts]  # Remove invalid values
df['Do_Not_Contact'] = contacts
df
```

```python
# Remove customers who should not be contacted or have missing phone numbers
df = df[df['Do_Not_Contact'] != 'Yes']
df = df[df['Do_Not_Contact'] != '']
df = df[df['Phone_Number'] != '']
df
```

```python
# Reset DataFrame index after filtering rows
df = df.reset_index(drop=True)
df
```

```python
# Remove duplicate records and drop unneeded columns
df = df.drop_duplicates()
df = df.drop('Address', axis=1)  # Drop the original 'Address' column
df = df.drop('zipcode', axis=1)  # Drop 'zipcode' column
df
```

```python
# Rename columns to proper names
df = df.rename(columns={'address': 'Address', 'state': 'State'})
df
```

```python
# EXPLORATORY DATA ANALYSIS
```

```python
# Distribution of 'Paying Customer'
plt.figure(figsize=(8, 6))
sns.countplot(x='Paying Customer', data=df)
plt.title('Paying vs Non-Paying Customers')
plt.show()
```

```python
# Distribution of phone number lengths
df['Phone_Number_Length'] = df['Phone_Number'].apply(len)
plt.figure(figsize=(8, 6))
sns.histplot(df['Phone_Number_Length'], kde=True)
plt.title('Distribution of Phone Number Lengths')
plt.show()
```
